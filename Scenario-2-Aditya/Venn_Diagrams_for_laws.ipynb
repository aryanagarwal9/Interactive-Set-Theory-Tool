# Venn Diagrams for Laws

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2_unweighted, venn3_unweighted
```

```python
A = set()
B = set()
C = set()
```

## A|B

```python
#aed7ff, #3598ff
diagram = venn2_unweighted([A,B], set_labels = ("A", "B"))
diagram.get_label_by_id("10").set_text("")
diagram.get_label_by_id("01").set_text("")
diagram.get_label_by_id("11").set_text("")
diagram.get_patch_by_id("10").set_color("#3598ff")
diagram.get_patch_by_id("01").set_color("#3598ff")
diagram.get_patch_by_id("11").set_color("#3598ff")
plt.title("A|B", x=0.5, y=-0.2, fontsize=20)
```
